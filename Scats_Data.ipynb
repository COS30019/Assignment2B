## Read Data

```python

df = pd.read_excel("/Users/mouyly/Downloads/jupiter_notebook/Scats Data October 2006.xls", sheet_name="Data", skiprows=6, header=None)
df.head()
```

```python
# STEP 3: Rename the columns
# There are 106 columns total → 11 metadata + 95 traffic columns
metadata_columns = ['Site ID', 'Location', 'Code', 'Group', 'Lat', 'Long', 'Extra1', 'Extra2', 'Extra3', 'Extra4', 'Date']
time_columns = [f'T{i}' for i in range(1, 96)]  # T1 to T95
df.columns = metadata_columns + time_columns
df.head()
```

```python
# Convert Excel float to actual date
df['Date'] = pd.to_datetime('1899-12-30') + pd.to_timedelta(df['Date'], unit='D')
df[['Site ID', 'Date']].head()
```

## Reshaping the data

```python
# STEP 4: Melt the wide format into long format
long_df = pd.melt(
    df,
    id_vars=['Site ID', 'Date'],          # Keep these columns
    value_vars=time_columns,              # Melt the 15-min traffic columns
    var_name='Interval',                  # The melted column name for time intervals
    value_name='Volume'                   # This will store the traffic counts
)
long_df.head()
```

```python
long_df['IntervalNum'] = long_df['Interval'].str.extract(r'T(\d+)').astype(int)
long_df['Time'] = pd.to_timedelta((long_df['IntervalNum'] - 1) * 15, unit='minutes')
long_df[['Interval', 'IntervalNum', 'Time']].head()
```

```python
# STEP 6: Extract interval number from 'T1', 'T2', ... and convert to time
long_df['Timestamp'] = long_df['Date'] + long_df['Time']
long_df[['Site ID', 'Timestamp', 'Volume']].head()
```

```python
long_df = long_df.sort_values(by=['Site ID', 'Timestamp'])
long_df[['Site ID', 'Timestamp', 'Volume']].head()
```

```python
# STEP 7: Combine date + time into a full timestamp
long_df['Timestamp'] = long_df['Date'] + long_df['Time']
long_df[['Timestamp', 'Volume']].head()
```

```python
# STEP 8: Final tidy and sort
long_df = long_df.sort_values(by=['Site ID', 'Timestamp'])
long_df[['Site ID', 'Timestamp', 'Volume']].head(10)
```
